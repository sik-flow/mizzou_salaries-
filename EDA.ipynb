```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
plt.style.use('ggplot')
#plt.xkcd();
```

```python
df = pd.read_csv('./data/cleaned.csv')
```

```python
df.head()
```

```python
fig = plt.figure(figsize = (20, 8))

ax1 = plt.subplot(121)
avg_salary_by_location = df[df['Hourly_Salary'] == 'salary'].groupby('Location')['Rate'].mean().values
avg_salary = df[df['Hourly_Salary'] == 'salary']['Rate'].mean()
plt.bar(np.arange(len(avg_salary_by_location)), avg_salary_by_location, color = '#32CD32')
plt.xticks(np.arange(len(avg_salary_by_location)), ['Columbia', 'KC', 'Rolla', 'STL'])
plt.axhline(avg_salary, color = 'k', label = f'Overall Average Salary: $79,565')
fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax1.yaxis.set_major_formatter(tick) 
plt.legend()
plt.title('Avg Salary for UM Employees by Location', fontsize = 22)

ax2 = plt.subplot(122)
avg_hourly_by_location = df[df['Hourly_Salary'] == 'hourly'].groupby('Location')['Rate'].mean().values
avg_hourly = df[df['Hourly_Salary'] == 'hourly']['Rate'].mean()
plt.bar(np.arange(len(avg_hourly_by_location)), avg_hourly_by_location, color = '#32CD32')
plt.xticks(np.arange(len(avg_hourly_by_location)), ['Columbia', 'KC', 'Rolla', 'STL'])
plt.axhline(avg_hourly, color = 'k', label = f'Overall Avg Hourly Rate: ${np.round(avg_hourly,2)}')
fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax2.yaxis.set_major_formatter(tick) 
plt.legend(loc = 'lower left')
plt.title('Avg Hourly Rate for UM Employees by Location')
```

```python
def guess_gender(first, middle):
    first_name_gender = d.get_gender(first)
    middle_name_gender = d.get_gender(middle)
    if (first_name_gender == 'female') or (first_name_gender == 'male'):
        return first_name_gender
    if (first_name_gender == 'mostly_female') & (middle_name_gender == 'mostly_male'):
        return ''
    if (first_name_gender == 'mostly_male') & (middle_name_gender == 'mostly_female'):
        return ''
    else:
        if (middle_name_gender == 'female') or (middle_name_gender == 'male'):
            return middle_name_gender
        if (middle_name_gender == 'mostly_female') or (middle_name_gender == 'mostly_male'):
            return middle_name_gender.replace('mostly_', '')
        else:
            return ''
```

```python
df['Gender_Guess'] = df.apply(
    lambda row: guess_gender(row['First_Name'], row['Middle_Name']), axis=1)
```

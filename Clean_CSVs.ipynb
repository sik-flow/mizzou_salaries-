```python
import pandas as pd
```

```python
df_columbia = pd.read_csv('./data/columbia.csv')
df_kc = pd.read_csv('./data/kc.csv')
df_rolla = pd.read_csv('./data/rolla.csv')
df_stl = pd.read_csv('./data/stl.csv')
```

```python
df_columbia.head()
```

```python
df = pd.concat([df_columbia, df_kc, df_rolla, df_stl])
```

```python
df.head()
```

```python
df.reset_index(inplace = True, drop = True)
```

```python
def hourly_salary(rate):
    if '.' in rate:
        return 'hourly'
    else:
        return 'salary'
```

```python
df['Hourly_Salary'] = df['Rate'].map(hourly_salary)
```

```python
df.head(15)
```

```python
def clean_rate(rate):
    rate = rate.replace(',', '')
    return float(rate)
```

```python
df['Rate'] = df['Rate'].map(clean_rate)
```

```python
df.head()
```

```python
df.to_csv('./Data/Cleaned.csv', index = False)
```
